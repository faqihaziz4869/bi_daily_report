```python
import pandas as pd
import glob
import math
import numpy as np
```

```python
search_order=pd.concat([pd.read_csv(f, dtype=str) for f in glob.glob("E:\REPORTING PARTNER IDX\PROJECT SPX\DATABASE SPX\SPX Order 2022\*.csv")], ignore_index = True)
search_order.loc[:,["Order Status"]]
search_order.drop_duplicates(subset=['Waybill No.'], keep='first', inplace = True)
search_order.rename(columns={"Sender":"Warehouse"},inplace=True)
search_order.rename(columns={"Destination":"District"},inplace=True)
search_order.rename(columns={"Destination City":"City"},inplace=True)
search_order.rename(columns={"Input Time":"Date"},inplace=True)
search_order.rename(columns={"Order No.":"Order ID"},inplace=True)
search_order.rename(columns={"Recipient Name":"Buyer Name"},inplace=True)
search_order.rename(columns={"Recipient Phone No.":"Buyer Phone"},inplace=True)
search_order.rename(columns={"Recipient Address":"Buyer Address"},inplace=True)
search_order.rename(columns={"Payment Type":"Metode Pembayaran"},inplace=True)
search_order.rename(columns={"Calculated Weight":"Weight."},inplace=True)
search_order.rename(columns={"Order Status":"Remarks Order"},inplace=True)
search_order.rename(columns={"POD Status":"REMARKS"},inplace=True)
#Filter order sourcenya: hanya munculkan yang Shopee Express
SPX=search_order[search_order["Order Source"]=="Shopee Express"]
SPX.drop(columns=['Order Source'], inplace=True)
```

```python
csvfile = "E:\REPORTING PARTNER IDX\PROJECT SPX\DATABASE SPX\Mapping Area SPX.csv"
mapping = pd.read_csv(csvfile, index_col = None)
search_order=search_order.merge(mapping,on=['City'],how='left')
```

```python
search_order=search_order.loc[:,["Waybill No.","Warehouse","Date","Order ID","District","City","Area","Pickup Time","Buyer Name","Buyer Phone","Buyer Address","Service Type","COD Amount","Metode Pembayaran","Weight.","Remarks Order","REMARKS"]]
```

```python
search_pod=pd.concat([pd.read_csv(f, dtype=str) for f in glob.glob("E:\REPORTING PARTNER IDX\PROJECT SPX\DATABASE SPX\SPX POD 2022/*.csv")], ignore_index = True)
search_pod=search_pod.sort_values(by = ["POD Scan Time"], ascending=[False])
search_pod.drop_duplicates(subset=["Waybill No."], keep='first', inplace = True)
search_pod.rename(columns={"POD Scan Time":"POD Time"},inplace=True)
search_pod.rename(columns={"Sign of Represented POD":"Remarks POD"},inplace=True)
search_pod=search_pod.loc[:,["Waybill No.","POD Time","Remarks POD","Signer"]]
main=search_order.merge(search_pod,how="left", on="Waybill No.")
main.info()
```

```python
return_center=pd.concat([pd.read_csv(f, dtype=str) for f in glob.glob("E:\REPORTING PARTNER IDX\PROJECT SPX\DATABASE SPX\SPX Return 2022/*.csv")], ignore_index = True)

return_center.rename(columns={'No. Waybill':'Waybill No.'},inplace=True)
return_center.rename(columns={'Cabang Asal':'Origin Branch'},inplace=True)
return_center.rename(columns={'Tipe Ekspress':'Express Type'},inplace=True)
return_center.rename(columns={'Tanggal Pengiriman':'Shipping Date'},inplace=True)
return_center.rename(columns={'Tipe Layanan':'Service Type'},inplace=True)
return_center.rename(columns={'Tipe Pembayaran':'Payment Type'},inplace=True)
return_center.rename(columns={'Nama Barang':'Item name'},inplace=True)
return_center.rename(columns={'Biaya Pengembalian':'ReturnShipping Fee'},inplace=True)
return_center.rename(columns={'Klien Pengirim':'Shipping Client'},inplace=True)
return_center.rename(columns={'Berat':'Weight'},inplace=True)
return_center.rename(columns={'Tujuan Pengiriman':'Return Destination'},inplace=True)
return_center.rename(columns={'Diregistrasi Oleh':'Registered by'},inplace=True)
return_center.rename(columns={'Waktu Pengembalian':'Return Time'},inplace=True)
return_center.rename(columns={'Cabang Terdaftar':'Register Branch'},inplace=True)
return_center.rename(columns={'Deskripsi':'Description'},inplace=True)
return_center.rename(columns={'Pusat Finansial Pengembalian':'Return Financial Center'},inplace=True)
return_center.rename(columns={'Status Pembatalan':'Void Status'},inplace=True)
return_center.rename(columns={'Waktu Pembatalan':'Void Time'},inplace=True)
return_center.rename(columns={'Dibatalkan Oleh':'Void by'},inplace=True)
return_center.rename(columns={'Cabang Pembatalan':'Void Branch'},inplace=True)
return_center.rename(columns={'Status Konfirmasi':'Confirm Status'},inplace=True)
return_center.rename(columns={'Dikonfirmasi Oleh':'Confirmed by'},inplace=True)
return_center.rename(columns={'Waktu Konfirmasi':'Confirm Time'},inplace=True)
return_center.rename(columns={'Cabang Konfirmasi':'Confirm Branch'},inplace=True)
return_center.rename(columns={'Penerima Pengembalian':'Return Recipient'},inplace=True)
return_center.rename(columns={'Waktu POD Pengembalian':'Return POD Time'},inplace=True)
return_center.rename(columns={'Cabang POD Pengembalian':'Return POD Branch'},inplace=True)
return_center.rename(columns={'POD Pengembalian':'Return POD'},inplace=True)
return_center.rename(columns={'Kurir Retur':'Return POD Courier'},inplace=True)
return_center.rename(columns={'Wali POD':'Represented POD'},inplace=True)
return_center.rename(columns={'Hubungan dengan Penerima':'Relationship with Recipient'},inplace=True)
return_center.rename(columns={'Sumber Order':'Order Source'},inplace=True)
return_center.rename(columns={'Lampiran':'Attachment'},inplace=True)
return_center.rename(columns={'Deskripsi Alasan Retur':'Return Reason Description'},inplace=True)

return_center.drop_duplicates(subset=["Waybill No."], keep='first', inplace = True)
return_center.rename(columns={"Confirm Status":"Returned"},inplace=True)
return_center=return_center.loc[:,["Waybill No.","Returned","Return POD"]]
result=main.merge(return_center,how="left", on="Waybill No.")
result
```

```python
result=main.merge(return_center,how="left", on="Waybill No.")
#main=main.loc[:,["Waybill No.","Warehouse","District","Date","Order ID","Buyer Name","Buyer Phone","Buyer Address","Metode Pembayaran","Weight.","Remarks Order"]]
result.loc[result["Remarks Order"]=="Picked Up", "POD Status"]="Delivery Process"
result.loc[(result["POD Status"]=="Delivery Process"),"REMARKS"]="Not Signed Yet"
result.loc[result["Remarks Order"]=="Cancel Order", "POD Status"]="Cancel Order"
result.loc[result["Remarks Order"]=="Schedule To Branch", "POD Status"]="Not Picked Up"
result.loc[result["Remarks Order"]=="Schedule To Courier", "POD Status"]="Not Picked Up"
result.loc[result["Remarks Order"]=="Pending", "POD Status"]="Not Picked Up"
result.loc[pd.notna(result["POD Time"]),"POD Status"]="Delivered"
#result.loc[(result["Remarks POD"]=="Signed"),"POD Status"]="Delivered"
#result.loc[pd.isna(result["Remarks POD"]),"POD Status"]="Delivery Process"
#result.loc[pd.isna(result["Pickup Time"]),"POD Status"]="Not Pickup"
result.loc[(result["Returned"]=="Confirmed"),"POD Status"]="Return Process"
result.loc[(result["Return POD"]=="Signed"),"POD Status"]="Returned"
result.loc[(result["Return POD"]=="Signed"),"REMARKS"]="Signed"
result.loc[(result["Remarks POD"]=="Signed"),"REMARKS"]="Signed"
result.loc[(result["POD Status"]=="Return Process"),"REMARKS"]="Not Signed Yet"
result.loc[(result["POD Status"]=="Not Picked Up"),"REMARKS"]="Not Signed Yet"
result.loc[(result["Remarks POD"]=="Signed"),"POD Status"]="Delivered"
result.loc[(result["POD Status"]=="Delivered"),"REMARKS"]="Signed"
#result.loc[pd="Picked Up"(result["Remarks Order"]),&pd.isna(result["Remarks POD"]),"POD Status"]="Delivery Process"
#result.loc[pd.isna(result["Final Status"])&pd.notna(search_waybill["Return Reason Description"])&pd.isna(search_waybill["Return POD Branch"]), "Final Status"] = "Return Process"
result
```
