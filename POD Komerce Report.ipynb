```python
import pandas as pd
import glob
import math
import numpy as np
```

```python
search_pod=pd.concat([pd.read_csv(f, dtype=str) for f in glob.glob("E:\REPORTING PARTNER IDX\PROJECT KOMERCE\POD Komerce/*.csv")], ignore_index = True)
search_pod["POD Recording Time"]=pd.to_datetime(search_pod["POD Recording Time"], format='%d-%m-%Y %H:%M:%S')
search_pod["Shipping Date"]=pd.to_datetime(search_pod["Shipping Date"], format='%d-%m-%Y %H:%M:%S')
#search_pos["Input Time"]=search_pos["Input Time"].dt.date
search_pod=search_pod.loc[(search_pod['POD Recording Time']>='2022/07/11') & (search_pod['POD Recording Time']<='2022/07/12')]
search_pod=search_pod.sort_values(by = ["POD Recording Time"])
search_pod.drop_duplicates(subset=["Waybill No."], keep='first', inplace = True)
```

```python
search_pod.rename(columns={"POD Recording Time":"POD Time"},inplace=True)
```

```python
search_pod['Shipping Time'] =search_pod["Shipping Date"]=pd.to_datetime(search_pod["Shipping Date"], format='%d-%m-%Y %H:%M:%S')
search_pod.head()
```

```python
search_pod
search_pod.info()
```

```python
search_pod=search_pod.loc[:,["Waybill No.","Waybill Source","Shipping Date","Shipping Time","POD Branch","POD Time"]]
komerce=search_pod[search_pod["Waybill Source"]=="Komerce"]
komerce.drop(columns=['Waybill Source'], inplace=True)
#komerce=komerce.loc[:,["Waybill No.","POD Recording Time"]]
komerce
```

```python
search_waybill=pd.concat([pd.read_csv(f, dtype=str) for f in glob.glob("E:\DATABASE\Waybill 2022/*.csv")], ignore_index = True)
search_waybill["POD Time"]=pd.to_datetime(search_waybill["POD Time"], format='%d-%m-%Y %H:%M:%S')
```

```python
search_waybill=search_waybill.loc[:,["Waybill No.","Origin Branch","Pickup Time","Destination","Item Name","Calculated Weight","COD Amount","Insurance Fee","Standard Shipping Fee","Receivable Shipping Fee","Handling Fee","Other Fee","Total Shipping Fee","VIP Username","Sender","Sender Handphone","Recipient","Recipient Handphone","Recipient Address","Payment Type","Express Type","POD Status","Waybill Source","Origin Province","Origin City","Destination Province","Destination City"]]
search_waybill.drop_duplicates(subset=["Waybill No."], keep='first', inplace = True)
result=komerce.merge(search_waybill,how="left", on="Waybill No.")
result
```
